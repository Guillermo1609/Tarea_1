```
def impuesto(sal):
 while (0>=sal):
   sal=int (input("El salario debe de ser mayor a 0, ingresa un nuevo salario:"))
 if (sal<=10000):
  print("El salario ingresado se encuentra en el rango de 0 & 10,000 pesos  por lo que tiene un descuento de impuesto del 5%")
  imp5=sal*0.05
  desc=sal-imp5
  return desc, True;
 elif (10000<sal<=15000):
  print("El salario  ingresado se encuentra en el rango de 10,000 & 15,000 pesos  por lo que tiene un descuento de impuesto del 10%")
  imp1=sal*0.1
  desc=sal-imp1
  return desc, True;
 elif (15000<sal<=20000):
  print("El salario ingresado se encuentra en el rango de 15,000 & 20,000 pesos por lo que tiene un descuento de impuesto del 15%")
  imp15=sal*0.15
  desc=sal-imp15
  return desc, True;
 elif (20000<sal):
  print("El salario ingresado se encuentra en el rango de 20,000 pesos en adelante por lo que tiene un descuento de impuesto del 18%")
  imp18=sal*0.18
  desc=sal-imp18
  return desc, True;

sal=float (input("Ingresa el salario mensual del trabajador:"))
print("El salario con el descuento es: ",impuesto(sal))
```
